# 手動輸入版

```python
import pandas as pd
import matplotlib.pyplot as plt
import re

data = pd.read_excel("資料集.xlsx")

mrt = []
for key, value in data.groupby(data["捷運站"]):
    mrt.append(key)


# 開場白
while True:
    print("饕客你好！歡迎使用「餐廳偵探」，我們收錄了超過30個捷運站，近1000家餐廳的資料，為你選擇今天最想吃的好評餐廳，並且…提供給你最真誠的評論！")



    while True:
        imrt = input("距離您最近的捷運站？輸入格式:站名，不加「站」字，例如「動物園」")  
        if imrt not in mrt:
            print("無此捷運站")
            continue

        else:
            print("您輸入的是",imrt,"捷運站")
            break

            

    print("你想吃的餐廳價位落在...")


    price = input("請輸入 $/$$/$$$/$$$$")

    print("您選擇的是",price,"價位")

    print("你想吃什麼類型的餐廳？")

    rest = input("請輸入 American/Chinese/Indian/Italian/Japenese/Mexican/Pizza/Bar/Steak/Cafe/Hot Pot/Seafood/Others/居酒屋/燒烤/韓式料理/餐酒館/歐風餐廳/東南亞餐廳/素食餐廳")

    print("您選擇的是",rest,"餐廳")


    output = data[data["價位($/$$/$$$)"] == price]
    output = output[output["MRTStation"] == imrt]
    output = output[output["餐廳分類(人工)"] == rest]
    if len(output["餐廳名稱"]) > 1:
        output = output.sample()

    if len(output["餐廳名稱"]) == 0:

        print("\t\t\t////糟糕!",imrt,"捷運站附近500公尺，查無這種餐廳。請重新選擇////")

        continue

    else:
        print("餐廳名稱:",list(output["餐廳名稱"])[0])
        print("餐廳價位:",price)
        print("兩年內餐廳星等平均值",list(output["兩年內星等（清除空值)"])[0])
        print("兩年內餐廳星等標準差",list(output["標準差"])[0])
        if int(output["標準差"]) > 1.5:
            print("這家餐廳正反風向差距有點大，要注意唷~~")


        #畫星等比例圖
        print("接下來看看各個星等分布比例")
        sl = [ float(output["比例"]) ,float(output["比例.1"]) ,float(output["比例.2"])  ,float(output["比例.3"])  ,float(output["比例.4"])]


        #情緒分析結果
        print("網友最在意這家餐廳的",list(output["最常出現 10 字詞(評論斷詞統計)"])[0][2:-2])
        print("隨機一句負評:",list(output["五星真誠好評"])[0])
        print("隨機一句好評:",list(output["一星真實差評"])[0])


        print("\t\t\t////今天的食物偵探就到此結束，如想找其他餐廳請重新訪問。////")

        break




```

# 按鈕版

```python
%matplotlib inline
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display
import pandas as pd

import re

data = pd.read_excel("資料集.xlsx")

mrt = []
for key, value in data.groupby(data["捷運站"]):
    mrt.append(key)

print("饕客你好！歡迎使用「餐廳偵探」，我們收錄了超過30個捷運站，近1000家餐廳的資料，為你選擇今天最想吃的好評餐廳，並且…提供給你最真誠的評論！")
print("請輸入基本資料")


C1 = widgets.Dropdown(
    options=mrt ,
    value = '大安',
    description='距離您最近的捷運站？',
    disabled=False,
)
display(C1)

C2 = widgets.Dropdown(
    options=["$","$$","$$$","$$$$"] ,
    value = '$$',
    description='你想吃的餐廳價位落在...',
    disabled=False,
)
display(C2)

C3 = widgets.Dropdown(
    options=['American','Chinese','Indian','Italian','Japenese','Mexican','Pizza','Bar','Steak','Cafe','Hot Pot','Seafood','Others','居酒屋','燒烤','韓式料理','餐酒館','歐風餐廳','東南亞餐廳','素食餐廳'],
    value = 'Chinese',
    description='你想吃什麼類型的餐廳？',
    disabled=False,
)
display(C3)


print("\n")

button = widgets.Button(description="填選完成")
out = widgets.Output()

display(button, out)

def on_button_clicked(b):
  with out:
    print("\n")
    msg1 = "您選擇的是 {} 捷運站".format(C1.value)
    print(msg1)
    msg2 = "您選擇的是 {} 價位".format(C2.value)
    print(msg2)
    msg3 = "您選擇的是 {} 餐廳".format(C3.value)
    print(msg3)
    print("\n")

    output = data[data["價位($/$$/$$$)"] == C2.value]
    output = output[output["MRTStation"] == C1.value]
    output = output[output["餐廳分類(人工)"] == C3.value]
    
    if len(output["餐廳名稱"]) > 1:
      output = output.sample()


    try:
      print("餐廳名稱:",list(output["餐廳名稱"])[0])
      print("\n")
      print("兩年內餐廳星等平均值",list(output["兩年內星等（清除空值)"])[0])
      print("兩年內餐廳星等標準差",list(output["標準差"])[0])
      if int(output["標準差"]) > 1.5:
          print("這家餐廳正反風向差距有點大，要注意唷~~")


      #畫星等比例圖
      print("\n")
      print("接下來看看各個星等分布比例")
      sl = [ float(output["比例"]) ,float(output["比例.1"]) ,float(output["比例.2"])  ,float(output["比例.3"])  ,float(output["比例.4"])]
      sx = [0,1,2,3,4]

      plt.plot(sl, color='blue', marker='o',mfc='red' )
      for x,y in zip(sx,sl):
          
          label = "{:.2f}".format(y)

          plt.annotate(label,
                      (x,y), 
                      textcoords="offset points", 
                      xytext=(0,10),
                      ha='center') 
      plt.xticks([0,1,2,3,4], ['1 star', '2 star', '3 star','4 star','5 star'])
      plt.show()

      #情緒分析結果
      print("\n網友最在意這家餐廳的",list(output["最常出現 10 字詞(評論斷詞統計)"])[0][2:-2])
      print("隨機一句好評:",list(output["五星真誠好評"])[0])
      print("隨機一句差評:",list(output["一星真實差評"])[0])


      print("\n\t\t\t////今天的食物偵探就到此結束，如想找其他餐廳請重新訪問。////")

    except IndexError:
      print("\t\t\t////糟糕!",C1.value,"捷運站附近500公尺，查無這種餐廳。請重新選擇////")


button.on_click(on_button_clicked)
```
